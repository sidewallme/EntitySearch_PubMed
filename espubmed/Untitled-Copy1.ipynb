```python
def extract(s, e1, e2):
    tags = normal(s)
    ents = entities(s)
    epos1 = []
    epos2 = []
    ret = []
    for i in range(0, len(ents)):
        if e1 in ents[i]:
            epos1.append(i)
        if e2 in ents[i]:
            epos2.append(i)
        
    for loc1 in epos1:
        for loc2 in epos2:
            if loc2 > loc1:
                ret.append("_".join(tags[loc1+1:loc2]))
    return ret               
```

```python
def normal(s):
    ret = []
    units = s.split(" ")
    for unit in units:
        ret.append(unit.split("|")[0])
    return ret
```

```python
def entities(s):
    ret = []
    units = s.split(" ")
    for unit in units:
        ret.append(unit.split("|")[1])
    return ret
```
